```python
# ==========================================
# MVP STARTUP - M칍DULOS DE MERCADEO Y LOG칈STICA
# ==========================================

# Variables globales (Listas para almacenar datos)
clientes_mercadeo = []
paquetes_logistica = []

# Men칰 Interactivo Principal
while True:
    print("\n" + "="*40)
    print("游 SISTEMA CENTRAL MVP 游")
    print("="*40)
    print("1. Registrar y Evaluar Clientes (Mercadeo)")
    print("2. Registrar y Clasificar Env칤os (Log칤stica)")
    print("3. Salir")
    
    opcion = input("Seleccione un m칩dulo (1, 2 o 3): ").strip()
    
    match opcion:
        # ---------------------------------------------------------
        # RETO 4: MERCADEO - "Fideliza-Puntos"
        # ---------------------------------------------------------
        case "1":
            print("\n--- M칩dulo de Mercadeo: Fidelizaci칩n ---")
            clientes_mercadeo.clear() # Limpiamos la lista por si se ejecuta varias veces
            
            # 1. Ingreso de 5 registros
            for i in range(5):
                print(f"\nCliente #{i+1}")
                nombre = input("Nombre del cliente: ").strip()
                compra = float(input("Valor total de compras acumuladas ($): "))
                clientes_mercadeo.append([nombre, compra])
            
            print("\n--- Reporte de Fidelizaci칩n ---")
            # 2. Ordenamiento alfab칠tico (.sort() ordenar치 por el primer elemento de la sublista, que es el nombre)
            clientes_mercadeo.sort()
            
            # 3. Procesamiento y L칩gica (for + if/else)
            for cliente in clientes_mercadeo:
                nombre_cliente = cliente[0]
                valor_compra = cliente[1]
                categoria = ""
                bono = 0
                
                # Asignaci칩n de categor칤as
                if valor_compra < 100000:
                    categoria = "Bronce"
                elif valor_compra <= 500000:
                    categoria = "Plata"
                else:
                    categoria = "Oro"
                    bono = valor_compra * 0.10 # Bono del 10%
                
                # Despliegue de resultados
                mensaje = f"Cliente: {nombre_cliente} | Compras: ${valor_compra:,.2f} | Categor칤a: {categoria}"
                if categoria == "Oro":
                    mensaje += f" | 游꾸 Bono de Regalo: ${bono:,.2f}"
                print(mensaje)
                
        # ---------------------------------------------------------
        # RETO 1: LOG칈STICA - "Ruta-칍ptima"
        # ---------------------------------------------------------
        case "2":
            print("\n--- M칩dulo de Log칤stica: Env칤os Nacionales ---")
            paquetes_logistica.clear()
            
            tarifa_base_kg = 4500 # Tarifa simulada por Kg
            
            # 1. Ingreso de 5 registros
            for i in range(5):
                print(f"\nPaquete #{i+1}")
                destino = input("Ciudad de destino (Ej. Medell칤n, Cali): ").strip()
                peso = float(input("Peso del paquete en Kg (Ej. 1.5): "))
                # Guardamos primero el peso para facilitar el ordenamiento
                paquetes_logistica.append([peso, destino])
                
            print("\n--- Manifiesto de Carga ---")
            # 2. Ordenamiento por peso (menor a mayor)
            paquetes_logistica.sort()
            
            # 3. Procesamiento y L칩gica (for + if/else)
            for paquete in paquetes_logistica:
                peso_paquete = paquete[0]
                destino_paquete = paquete[1]
                tipo_carga = ""
                
                # Clasificaci칩n
                if peso_paquete < 1:
                    tipo_carga = "Documento"
                elif peso_paquete <= 10:
                    tipo_carga = "Paqueter칤a"
                else:
                    tipo_carga = "Carga"
                
                # Costo de env칤o
                costo_envio = peso_paquete * tarifa_base_kg
                
                print(f"Destino: {destino_paquete} | Peso: {peso_paquete}kg | Tipo: {tipo_carga} | Costo Env칤o: ${costo_envio:,.2f}")

        # ---------------------------------------------------------
        # SALIDA DEL SISTEMA
        # ---------------------------------------------------------
        case "3":
            print("Saliendo del sistema MVP... 춰Mucho 칠xito con el emprendimiento!")
            break
            
        case _:
            print("Opci칩n no v치lida. Intente de nuevo.")
```
